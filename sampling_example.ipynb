```python
import sys
import os

from utils import search_subdir
import json 
import random
import math

from tqdm import tqdm

# load path
datasets_path = search_subdir('data')
```

```python
def list_to_str(context: list) -> str:
    context_str = ''
    for sen in context:
        speaker, sentence = next(iter(sen.items()))
        context_str += speaker.strip() + ':' + sentence.strip() + '\n'
    return context_str
```

## Sampling for one file

```python
# sampling
context_list = []
sample_num = 5
with open(datasets_path[0], encoding='utf-8') as f:
    data_json = json.load(f)
    sample_json = random.sample(data_json['data'], sample_num)
    context_list += [list_to_str(context['context']) for context in sample_json]
```

```python
context_list
```

```python
# save
with open("sample_data.txt", "w") as f:
    for call in context_list:
        f.write(call+'\n\n')
```

## Sampling for files

```python
total_count = 0
context_dic = {}
for dataset_path in tqdm(datasets_path):
    with open(dataset_path, encoding='utf-8') as f:
        data_json = json.load(f)
        context_dic[dataset_path] = data_json['data']
        total_count += len(data_json['data'])
```

```python
# Sampling to the number of data in the file
sample_num = 30

context_list = []
for dataset_path in tqdm(datasets_path):
        # Number of samples for dataset
        k = math.ceil(len(context_dic[dataset_path])/total_count * sample_num)
        
        sample_json = random.sample(data_json['data'], k)
        context_list += [list_to_str(context['context']) for context in sample_json]
```
