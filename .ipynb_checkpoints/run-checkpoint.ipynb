```python
from api import RequestChatGPT
from prompt import prompt_fn
from postprocessor import postprocessor_fn
```

```python
with open("sample_call.txt", "r") as f:
    text = ''.join(f.readlines())
```

```python
dialog_list = text.split('\n\n')
```

```python
len(dialog_list)
```

```python
key = "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"
api = RequestChatGPT(key)
```

```python
model_list = ['gpt-3.5-turbo', 'text-davinci-003', 'text-babbage-001']
```

```python
import pandas as pd

df = pd.DataFrame({'prompt_type':[], 'dialog':[], 'gpt-3.5-turbo':[], 'text-davinci-003':[], 'text-babbage-001':[]})
```

```python
for prompt_name in ["simple", "tuning1", "tuning2"]:
    for dialog in dialog_list:
        row_dic = {'prompt_type':prompt_name, 'dialog':dialog}
        for model_name in model_list:
            result = api.summary(model_name, prompt_fn[prompt_name](dialog))
            result = postprocessor_fn[prompt_name](result)
            row_dic[model_name] = result
        df = df.append(row_dic, ignore_index=True)
```

```python
df.to_csv('result.csv', index=False, sep='\t')
```
